```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

dates = pd.date_range(start='2024-11-20', periods=10, freq='D')

sp500_open = [4000.0, 3980.0, 3960.0, 3950.0, 3940.0, 3930.0, 3920.0, 3910.0, 3900.0, 3890.0]
sp500_close = [4020.0, 3995.0, 3985.0, 3965.0, 3950.0, 3940.0, 3935.0, 3925.0, 3915.0, 3905.0]
sp500_volume = [2800000, 2750000, 2600000, 2500000, 2400000, 2350000, 2300000, 2250000, 2200000, 2150000]

ftse100_open = [7300.0, 7280.0, 7250.0, 7240.0, 7230.0, 7220.0, 7210.0, 7200.0, 7190.0, 7180.0]
ftse100_close = [7305.0, 7295.0, 7280.0, 7265.0, 7255.0, 7240.0, 7230.0, 7215.0, 7210.0, 7195.0]
ftse100_volume = [1500000, 1450000, 1400000, 1350000, 1300000, 1250000, 1200000, 1150000, 1100000, 1050000]

sp500_df = pd.DataFrame({
    'Date': dates,
    'Open': sp500_open,
    'Close': sp500_close,
    'Volume': sp500_volume
})

ftse100_df = pd.DataFrame({
    'Date': dates,
    'Open': ftse100_open,
    'Close': ftse100_close,
    'Volume': ftse100_volume
})

plt.figure(figsize=(10, 6))

plt.subplot(2, 1, 1)
plt.plot(sp500_df['Date'], sp500_df['Close'], marker='o', color='b', label='S&P 500 Close')
plt.title('S&P 500 Index - Closing Prices')
plt.xlabel('Date')
plt.ylabel('Price')
plt.legend()


plt.subplot(2, 1, 2)
plt.plot(ftse100_df['Date'], ftse100_df['Close'], marker='x', color='r', label='FTSE 100 Close')
plt.title('FTSE 100 Index - Closing Prices')
plt.xlabel('Date')
plt.ylabel('Price')
plt.legend()

plt.tight_layout()
plt.show()


plt.figure(figsize=(10, 6))
plt.plot(sp500_df['Date'], sp500_df['Close'], marker='o', color='b', label='S&P 500 Close')
plt.plot(ftse100_df['Date'], ftse100_df['Close'], marker='x', color='r', label='FTSE 100 Close')
plt.title('S&P 500 vs FTSE 100 - Closing Prices')
plt.xlabel('Date')
plt.ylabel('Price')
plt.legend()
plt.show()

plt.figure(figsize=(10, 6))
plt.scatter(sp500_df['Close'], ftse100_df['Close'], color='purple')
plt.title('Scatter Plot - S&P 500 vs FTSE 100')
plt.xlabel('S&P 500 Closing Prices')
plt.ylabel('FTSE 100 Closing Prices')
plt.show()


width = 0.35  # width of bars
plt.figure(figsize=(10, 6))
bar1 = plt.bar(sp500_df['Date'] - pd.Timedelta(days=0.2), sp500_df['Volume'], width=width, label='S&P 500 Volume', color='b')
bar2 = plt.bar(ftse100_df['Date'] + pd.Timedelta(days=0.2), ftse100_df['Volume'], width=width, label='FTSE 100 Volume', color='r')

plt.title('Volume Comparison - S&P 500 vs FTSE 100')
plt.xlabel('Date')
plt.ylabel('Volume')
plt.legend()
plt.show()
```
